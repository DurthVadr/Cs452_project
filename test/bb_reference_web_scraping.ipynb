```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the advanced stats page
url = 'https://www.basketball-reference.com/leagues/NBA_2025_advanced.html'

# Fetch the content from the URL
response = requests.get(url)
content = response.content

# Parse the HTML content
soup = BeautifulSoup(content, 'html.parser')

# Find the table containing advanced stats
body = soup.find('body', {'class': 'bbr'})
wrap = body.find('div', {'id': 'wrap'})
content = wrap.find('div', {'id': 'content', 'role': 'main'})

all_advanced_team = content.find('div', { 'id': 'all_advanced_team', 'class': 'table_wrapper'})
div_advanced_team = all_advanced_team.find('div', {'id': 'div_advanced-team'})
table = div_advanced_team.find('table', {'class': 'stats_table sortable now_sortable is_sorted', 'id': 'advanced_stats'})

headers = table.find('thead').find_all('th')
headers = [ele.text.strip() for ele in headers]


rows = table.find('tbody').find_all('tr')
data = []
for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    data.append([ele for ele in cols if ele])




# Create a DataFrame
advanced_stats_df = pd.DataFrame(rows, columns=headers[1:])  # Skip the first header as it is the index

# Save the DataFrame to a CSV file
advanced_stats_df.to_csv('advanced_stats_nba_2025.csv', index=False)
```
